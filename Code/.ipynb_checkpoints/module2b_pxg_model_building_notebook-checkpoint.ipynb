### Main Function which triggers all function from Helper library module 

```python
def run_model_function(df, flag, travel_flag, id_vars, target_variable, split_train, treatment_flag, mo_name, models_path):
    
    '''
    Purpose : Prepare data for modeling, divide into training and testing sets (50-50),
              Triggers Grid search, Saves the best model, Runs calibration for assessment if 
              isotonic/sigmoid reduces the mean square error
    
    Arguments : 1) df              : Pandas dataframe for an individual treatment (created in code)
                2) flag            : Model flag  which is used for identifying the ds related to treatment (defined in code)
                3) travel_flag     : Travel flag which keeps only travelers in the modeling dataset (defined in code)
                4) id_vars         : id variables which we want to keep but not in modeling (defined in code)
                5) target_variable : Target variable in Df (defined in code)
                6) split_train     : Ratio of spliting (defined in code)
                7) treatment_flag  : Redundant variable (kindly ignore)
                8) mo_name         : Defined by us for each treatment (defined in code)
                9) models_path     : Defines the location where model objects will be dumped (Defined in JSON file)
    
    Returns   : Saves model objects and print various outputs
    
    '''   
  

    print "Pipeline for Below Model has started running"
    print "Checkpoint 1: Keeping only Travelers and then Dividing Target, ID and Features dataframes"
    
    df_travel = df[df[travel_flag]==1]
    
    target_var = [target_variable]
    df_target  = df_travel[target_var]   #Defining target DF
    df_id      = df_travel[user_id_vars] # Defining ID Dataframe

    # Drop target and ID features from the dataframe to obtain final features dataframe
    
    df_features = df_travel.drop(user_id_vars,axis=1)
    df_features_final = df_features.drop(target_var,axis=1)  

    print "Target dataset created:", df_target.shape
    print "ID Dataset created for reference:", df_id.shape
    print "Features dataset shape",  df_features_final.shape

    df_target['column_strat'] = df_target[target_variable].astype(str) + df_id['segment'].astype(str)


    print "Checkpoint 2: Calling Function creating_test_train_sampled_datasets"
    xtrain, xtest, ytrain, ytest , idtrain, idtest =   creating_test_train_sample_seeded_ds(df_target, df_id, \
                                                       df_features_final, 1000, split_train, target_variable)

    print "Shape of training dataset            :", xtrain.shape
    print "Shape of validation dataset          :", xtest.shape
    
    
    parameters = dict(
                  feature_selection__k=[35,40,45],
                  feature_selection__score_func=[f_classif],     
                  random_forest__n_estimators=[100,150,200,250,500],
                  random_forest__min_samples_split=[5,10,15,50,100,200,300,400,500]
                  )
        
    
    # Creating path to dump model object
    MODEL_GRID_OUTPUT_FILE = models_path +"/"+ mo_name + flag
    
    print "Checkpoint 3: Calling Function trig_gs_rf"
    model_object = trig_gs_rf(xtrain, ytrain, parameters, flag, target_variable)
    
    # Dumping Model Object
    joblib.dump(model_object,MODEL_GRID_OUTPUT_FILE,compress=1)
    
    # Loading Back Model Object
    loaded_model = joblib.load(MODEL_GRID_OUTPUT_FILE)
    
    print(loaded_model.best_estimator_)
    print(loaded_model.best_params_)
    
    print "Checkpoint 4: Calling Function print_model_stats"
    globals()["feature_importance"+flag] = print_model_stats(loaded_model.best_estimator_, xtrain, ytrain, xtest, ytest, target_variable)
    #plt.gcf().clear()
    
    # Plot calibration curve for Random Forrest Classifier
    
    isotonic = CalibratedClassifierCV(loaded_model.best_estimator_, cv=10, method='isotonic')
    sigmoid = CalibratedClassifierCV(loaded_model.best_estimator_, cv=10, method='sigmoid')
    
    print "Checkpoint 5: Calling Function calibration_curve_and_classifiers"
    calibration_curve_and_classifiers(xtrain, ytrain, xtest, ytest, loaded_model.best_estimator_, "Random Forrest", isotonic, sigmoid)
    #plt.gcf().clear()
    
    # Creating Global Datasets in order to debug if required and also used to generate files for features asessment
    
    globals()["xtrain"+flag]=xtrain
    globals()["xtest"+flag]=xtest
    globals()["ytrain"+flag]=ytrain
    globals()["ytest"+flag]=ytest
    globals()["idtrain"+flag]=idtrain
    globals()["idtest"+flag]=idtest

    print "--------------------------------------------------------------"
    print "--------------------------------------------------------------"   
```
